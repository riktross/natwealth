The central bank interest rate is an economic variable of critical importance in the economic development of a nation. It determines the returns of all government securities, and every investor uses it as a benchmark when making risky investment decisions. No investment will be made unless its expected rate of return is higher than the central bank interest rate by a margin large enough to compensate for the riskyness of the investment. Put simply, **increasing the interest rate decreases the viability of real investments in the economy.**

Brazilian interest rates are incredibly high, and have been a drag on investments for a long time. In order to put this into perspective, I have plotted the nominal returns in the Brazilian stock market (Ibovespa) versus the Brazilian central bank interest rate (Selic). The comparison can give an idea of how profitable investing in real economic activity versus the risk-free government subsidy. Over a sufficiently long time period of time, the stock market *should* outperform the central bank rate, as investors require compensation for the higher risk of stock market positions, as any introductory finance text book will tell you. The intuition is simple, higher risk -> higher reward.

It turns out that in Brazil, the opposite is the case: **the risk-free investment outperforms the risky investment!** The first plot below shows the development of a R$100 investment made in August 1994 (the early months after the Brazilian Real was adopted as currency) up to August 2023. After 29 years the investment in the security that tracks the central bank interest rate grew to ~6800 Reais while the stock market investment grew to ~2200 Reais. The total return of the investment was 3 times that of the risky stock market investment!

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

from pathlib import Path

project_data_folder = Path(".").resolve().parents[0] / "data" / "2023_bovespa_v_selic"

# Load data selic
data_selic_raw = pd.read_csv(project_data_folder / "STP-20230826131322637.csv", 
                             sep=";")

data_selic = data_selic_raw.copy()
data_selic.columns = ["ts", "rate"]

# Skip last entry since it does not contain interest data (holds metadata)
data_selic = data_selic.iloc[0:-1, :]

# Properly format the timestamp column
data_selic["ts"] = pd.to_datetime(data_selic["ts"], format="%d/%m/%Y")

# Transform SELIC column to numeric
## dot represents thousands, remove
data_selic["rate"] = data_selic["rate"].str.replace(".", "", regex=False)
## comma represents decimal, change to dot
data_selic["rate"] = data_selic["rate"].str.replace(",", ".", regex=False)
## change to numeric
data_selic["rate"] = pd.to_numeric(data_selic["rate"])

# selic interest factor, daily instead of annualized
data_selic["factor"] = (1+(data_selic["rate"]/100))**(1/252)

# choose starting date of 4th of July 1994
data_selic = data_selic.loc[data_selic["ts"] >= dt.datetime(1994, 8, 25), :]

data_selic["index"] = data_selic["factor"].cumprod()

# normalize index to be at 100 at the start
data_selic["index"] = (data_selic["index"]/data_selic["index"].iloc[0])*100

data_selic = data_selic[["ts", "index"]]
data_selic = data_selic.rename(columns={"index": "index_selic"})

# Load data Bovespa
data_bvsp_raw = pd.read_csv(project_data_folder/"^BVSP.csv")

# read only relevant columns
data_bvsp = data_bvsp_raw[["Date", "Adj Close"]].copy()

# rename columns
data_bvsp.columns = ["ts", "index"]

# properly format time series
data_bvsp["ts"] = pd.to_datetime(data_bvsp["ts"], format="%Y-%m-%d")

# start in 4th of july 1994
data_bvsp = data_bvsp.loc[data_bvsp["ts"] >= dt.datetime(1994, 8, 25)]

# normalize index to be at 100 at the start
data_bvsp["index"] = (data_bvsp["index"]/data_bvsp["index"].iloc[0])*100

data_bvsp = data_bvsp.rename(columns={"index": "index_bovespa"})

# merge both indices
data = pd.merge(data_bvsp, data_selic, "outer", on="ts").sort_values("ts")

# Fill in all days
all_days = pd.DataFrame({"ts":pd.date_range(data.ts.min(), data.ts.max(), freq="D")})
data = pd.merge(data, all_days, "outer", on="ts").sort_values("ts")

data = data.ffill()

data = data.set_index("ts")
```

```python
%matplotlib widget
# Matplotlib line plot using subplot
fig, ax = plt.subplots(dpi=150)

ax.plot(data.index, data['index_bovespa'], label='Bovespa')
ax.plot(data.index, data['index_selic'], label='Selic')

# Axis labels and title
ax.set(xlabel='Date', ylabel='Index (Aug 1994 = 100)', title='Selic and Bovespa cumulative returns')



# Add light horizontal grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Legend
ax.legend()

# Show the plot
plt.show()
```

This second plot shows the rolling annual rate of return of the two investments, highlighting the risk differential. I know that I would prefer to enjoy the smooth ride on the orange line (earning 3 times as much) than the bumpy ride of the bovespa)

```python
def get_returns(series):
    return ((series.iloc[-1]/series.iloc[0]) - 1)*100

data_rolling_returns = data.rolling(365).apply(get_returns).dropna()


# Matplotlib line plot using subplot
fig, ax = plt.subplots(dpi=150)

ax.plot(data_rolling_returns.index, data_rolling_returns['index_bovespa'], label='Bovespa')
ax.plot(data_rolling_returns.index, data_rolling_returns['index_selic'], label='Selic')

# Axis labels and title
ax.set(xlabel='Date', ylabel='Annual rate of return (%)', title='Selic and Bovespa rolling annual returns')

# Add light horizontal grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Legend
ax.legend()

# Show the plot
plt.show()
```

It is not surprising that depending on the measure, studies often estimate a negative equity risk premium for the Brazilian economy, such as [Walter Gonalves, et al (2011)](https://www.proquest.com/openview/3ca1cba8c4d6cf0478b751d98367201c/1?pq-origsite=gscholar&cbl=2028896).

The problem lies not in the bad performance of the stock market, but in the absurdly high interest rate. Over the entire period, the Stock market return has been around 11.1% per year (a normal rate of return), while the central bank rate has been around 15.7% per year.

So finally, the remaing question is: **Why would any investor in this country actually want to invest real productive activities instead of earning a rent from the central bank?**

The absurdity of the past 30 years is undeniable. As of writing, Brazil still has one of the highest real rates of interest in the world ([Second place in this list of 40 countries](https://g1.globo.com/economia/noticia/2023/12/13/ranking-mundial-taxa-de-juros-reais.ghtml)) and I do not have high hopes a structural shift any time soon. Why are they so high? Beyond being a stable source of income for rentiers, the high rate keeps the Real appreciated (by keeping demand for Brazilian currency high for foreign investors) and thus sustains the real purchasing power of Brazilians that have to buy industrial goods produced abroad. Lowering the interest rate significantly and permanently would not be a policy decision free of costs and risks.
